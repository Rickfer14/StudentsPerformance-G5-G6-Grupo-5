# ANALIZAR LA CORRELACIÓN ENTRE EL NIVEL EDUCATIVO PARENTAL Y LA PREPARACIÓN PREVIA PARA EXÁMENES EN ESTUDIANTES DE 5º Y 6º GRADO, CONSIDERANDO EL SEXO DEL ESTUDIANTE COMO FACTOR CLAVE EN ESTUDIANTES CON PADRES DE MENOR FORMACION EDUCATIVA

## **EL PROBLEMA**
## Planteamiento del problema
La educación es fundamental para el desarrollo individual y social, y su calidad está influenciada por múltiples factores, como el entorno familiar y las condiciones socioeconómicas, esto según la OECD, en el año 2018. Además, estudios señalan que el nivel educativo de los padres es un predictor clave en el aprendizaje y rendimiento de los niños en sus estudios, ya que familias con mayor educación suelen destinar recursos y tiempo a mejorar las estrategias de aprendizaje, tal como señala Crosnoe y otros, en el año 2010.

No obstante, pueden entrar en juego otro tipo de factores, como el sexo del estudiante, lo que podría determinar este desempeño académico de los jóvenes (Fundación CYD, 2023).

La preparación previa académica es un elemento fundamental, siendo esta, la llave que traslada al estudiante al éxito académico. Caracterizándose por aplicar métodos de estudio y adherir buenas prácticas, es, sin duda, la mejor manera de conseguir este éxito anteriormente mencionado.

En el contexto de educación primaria, esta preparación previa para exámenes persiste como factor clave para la obtención de un buen rendimiento académico (Hattie, 2009). 

Sabiendo que el sexo del estudiante puede actuar como determinante en el rendimiento académico, esto ayuda a analizar comportamientos de ciertos tipos de niveles educativos parentales, lo que da pie a la siguiente pregunta:  


## Formulación del Problema
¿Existe una correlación significativa entre el nivel educativo de los padres y la preparación previa para exámenes en estudiantes de 5to y 6to grado, considerando el sexo del estudiante como factor clave en estudiantes con padres de menor formación educativa?


## Interrogantes de la Investigación
- ¿Cómo se distribuyen las categorías de preparación para exámenes (completado/Ninguno) entre los diferentes niveles educativos parentales?

- ¿Los estudiantes con padres de mayor nivel educativo tienen una tasa de preparación para exámenes significativamente mayor que aquellos con padres de menor formación?

- ¿La preparación previa para exámenes tiene un impacto significativo en las notas de los estudiantes, según el nivel educativo parental? 

- ¿El sexo del estudiante influye en la preparación para exámenes en estudiantes con padres de menor nivel educativo (Algún estudio de secundaria)?


## **OBJETIVOS DE LA INVESTIGACIÓN**

## Objetivo General
-Analizar la Correlación Entre el Nivel Educativo Parental y la Preparación Previa para Exámenes en Estudiantes de 5to Y 6to Grado. Considerando el sexo del estudiante como factor clave en estudiantes con padres de menor formación educativa.

## Objetivos Específicos
-Distribuir las categorías de preparación para exámenes (completado/Ninguno) entre los diferentes niveles educativos parentales.

-Comparar si los estudiantes con padres de mayor nivel educativo tienen una tasa de preparación para exámenes significativamente mayor que aquellos con padres de menor formación.

-Evaluar el impacto de la preparación previa para exámenes en las notas de los estudiantes, según el nivel educativo parental.

-Analizar si el sexo del estudiante influye en la preparación para exámenes en estudiantes con padres de menor nivel educativo “Algún estudio de secundaria” (caso atípico).


## **JUSTIFICACIÓN DE LA INVESTIGACIÓN**
La preparación previa para los exámenes es, claramente, un predictor clave sobre el desempeño de los estudiantes en el ámbito académico. Sin embargo, dependiendo de la calidad de estudios que se presenten en los padres de los estudiantes, dicha preparación puede verse afectada cuando el nivel educativo parental es bajo. Este problema se agrava aún más en contextos donde el nivel educativo parental es bajo y el sexo del estudiante puede influir en las expectativas de rendimiento.

Al realizar este estudio, se busca validar la problemática principal planteada, sirviendo como base para analizar las desigualdades educativas en casos donde el estudiante no tiene cierta influencia parental, y así optimizar la toma de decisiones en instituciones educativas donde se presenten estos casos.

Como estudios previos han explorado el rol del nivel educativo parental en la educación, es por ello que esta investigación busca, además de enfocar como los diferentes grupos de los niveles educativos parentales de los estudiantes influyen de manera diferente en la realización de la preparación previa para los exámenes, como influye también el sexo del estudiante, en donde los niveles educativos parentales no son tan altos.



## **LÍMITES Y ALCANCES**

## Límites
El estudio está sustentado en una base de datos extraída de internet, por lo que no se puede generalizar a una población real.

Los resultados pueden no ser equivalentes a los obtenidos de una distribución de estudiantes donde se recolecten las mismas características.


## Alcances
Proporciona un análisis descriptivo replicable para analizar las desigualdades educativas.

Se focaliza el análisis en un grupo específico: estudiantes con padres de alta y baja formación educativa.


## **MARCO TEÓRICO**

## Antecedentes De La Investigación
Inicialmente se acudió a la investigación de Miranda y otros en el año 2014, quienes realizaron una investigación en la Pontificia Universidad Católica del Perú, titulada: “La competencia parental y el nivel de aprendizaje en estudiantes de 5to a 6to grado de primaria”. El presente estudio, tuvo como propósito principal, conocer e identificar la efectividad de la evaluación de la Competencia Parental y el nivel de aprendizaje en estudiantes de 5to A 6to grado de primaria, de una institución educativa de gestión estatal del distrito de Independencia. Para la recolección de la información se utilizó la prueba “Evaluación de la Competencia Parental” (versión hijos/as), para medir hasta qué punto los padres de familia influyen en el nivel de aprendizaje de sus hijos/as. En el proceso de análisis de los datos para determinar si existe relación entre la competencia parental y rendimiento escolar, se utilizó la prueba estadística coeficiente de correlación de Pearson. Los resultados obtenidos fueron: No existe correlación entre la Competencia parental percibida y el nivel aprendizaje en los estudiantes de 5to de primaria, mientras que, en referencia los estudiantes de 6to. grado se acepta la hipótesis de investigación: Si existe correlación entre la Competencia parental percibida y el aprendizaje.

Seguido por las investigadoras Espejel y García en el año 2019, las cuales basaron su investigación en el Instituto Politécnico Nacional, México, titulada: “Nivel educativo y ocupación de los padres: Su influencia en el rendimiento académico de estudiantes universitarios”. El objetivo de la presente investigación fue analizar la influencia del nivel educativo y la ocupación de los padres en el rendimiento académico de los estudiantes universitarios y así proponer estrategias que permitan elevar el logro académico. Los datos fueron obtenidos mediante un cuestionario semiestructurado y analizados con el algoritmo chi-squared automatic interaction detection (Chaid), tablas de contingencia y una regresión ordinal. Los resultados muestran que el nivel educativo de la madre, cuando es superior a licenciatura, influye positivamente en el rendimiento académico de los hijos; en cambio el nivel educativo del padre no resultó significativo estadísticamente. Se observó también que estudiantes cuyos padres se desempeñan como profesionistas, comerciantes, directivos, jefes de departamento y trabajadores de la educación son aquellos que obtienen un mejor rendimiento académico; la ocupación de la madre no resultó estadísticamente significativa.

## Bases Teóricas

### ***Nivel educativo parental***
Como indica Fiveable, el nivel educativo parental es el nivel de escolaridad formal y de logros educativos alcanzados por los padres. Además, dicho nivel educativo puede influir de ciertas maneras en los hijos, como en sus expectativas, acceso a recursos o procesos de estudio.

### ***Preparación previa para exámenes***
Es el proceso de preparación previa para rendir en alguna prueba, ya sea académica, deportiva o cualquier otro tipo. Esto se logra con el repaso y practica de temas o técnicas que vayan en dicha prueba.

### ***Nivel socioeconómico parental***
Es la combinación de factores económicos y sociales que presentan los padres de familia, tal como indica Merchán y otros (2022).

### ***Apoyo parental en educación***
Es el acompañamiento que brindan los padres en el proceso educativo de sus hijos, algo fundamental para lograr los objetivos académicos, y además mejora la autoestima estudiantil.

### ***Rendimiento académico***
Son los resultados educativos de los estudiantes, medible en exámenes o calificaciones. Se puede llegar a ella a través de los métodos mencionados anteriormente (como el apoyo parental en educación y la preparación previa para examenes).

### ***Capital cultural***
Pierre Bourdieu (1986) plantea que el capital cultural (conocimientos, hábitos y creencias) es transferible de generación en generación, influyendo también en las oportunidades académicas de los descendientes. Esto apoya la idea de que, familias con mayor educación, priorizan entornos educativos, proporcionando tutorías o materiales de estudio, asegurando el éxito escolar. Mientras, en familias con menos calidad educativa, lo que en mayoría de casos también conlleva a situación socioeconómica, se priorizan otros gastos, dejando de lado la preparación extracurricular.

### ***Teoría de la Autodeterminación***
Deci y Ryan (1985) proponen que la motivación interna (interés genuino por aprender) y la motivación externa (alguna recompensa a obtener) determinan el compromiso académico. Los estudiantes que poseen padres con una educación parental notable, tienden a desarrollar más esta motivación interna, mientras que estudiantes cuyo nivel académico de los padres no es tan elevado, desarrollan más esta motivación externa (como la obtención de alguna beca de estudio). Esto, aunado a los pocos hábitos académicos heredados de sus padres (además de la priorización a otro tipo de gastos, dejando de lado la inversión en recursos estudiantiles).

### ***Diferencias de género en la preparación académica***
En la educación infantil, los estudiantes no suelen tener diferencias significativas en su rendimiento académico según su sexo. Sin embargo, se vio con una pequeña mejor preparación a los hombres, esto según un estudio realizado por Medina y otros, en 2020.

## Explicación de las variables
### ***“Nivel educativo de los padres” y “Preparación”***
Se decide correlacionar estas dos variables para verificar si se interceptan. Es decir, si a mayor nivel educativo parental, los estudiantes tienden más a tener preparación previa para los exámenes, que es en lo que se basa este trabajo investigativo.

### ***“sexo”***
Se utilizará para el análisis de algún caso atípico, donde algunos de los valores de la variable “parental level of education” no sigan el comportamiento esperado.

### ***"Notas de los estudiantes"***
Se utilizará para demostrar, por grupos de nivel educativo parental dado que hayan completado la preparación previa para los exámenes, si dicha preparación es efectiva, comparando las notas de aquellos que la completaron con los que no.
